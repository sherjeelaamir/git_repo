```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt


# load Dataset
phool = sns.load_dataset("iris")
phool

 



```

```python
# Draw a lint plot
sns.lineplot(x="sepal_length" , y="sepal_width", data = phool)
plt.title("Phool Plots")
plt.show()
```

```python
sns.lineplot(x="sepal_length" , y="sepal_width", data = phool)
plt.title("Phool Plots")
plt.xlim(2)
plt.ylim(1)
sns.set_style(style = None, rc = None)
plt.show()
```
