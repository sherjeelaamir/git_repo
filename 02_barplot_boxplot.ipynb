```python
# horizontal plot
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy

# load Dataset
kashti = sns.load_dataset("titanic")
kashti

# Draw a bar plot with order and palette
sns.barplot(x="fare" , y="class", hue = "embarked" ,data = kashti ,
            linewidth = 0.1, #(o is thinnest )
            facecolor = (0.4,0.1,0.2,0.7), 
            errcolor = "0.5",
            edgecolor = "0", # 0 is dark and 1 is transparent
           palette = "dark", saturation = 1)
plt.title("Data Plot")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

kashti = sns.load_dataset("titanic")
kashti.head()

sns.boxplot(x="survived",
           y = "age", showmeans = True,
                meanprops = {"marker": "x",
                            "markersize": "12",
                            "markeredgecolor" : "red"},
           data = kashti)

#show labels

plt.ylabel("Age (Years)", size = 12),
plt.title("Box Plot", size = 18, weight = "bold"),
plt.xlabel("How many survived", size =13)
plt.show()
```
